```python
try:
    # Assuming the link to fixtures is found using a button or a navigation item
    fixtures_link = driver.find_element(By.XPATH, "//a[contains(@href, '/fixtures')]")  
    fixtures_link.click()
except NoSuchElementException:
    print("Fixtures link not found")
    driver.quit()

    # Step 3: Wait for the fixtures page to load
time.sleep(5)
```

# Question - 3. Scrape the details of State-wise GDP of India from statisticstime.com.
Url = http://statisticstimes.com/
You have to find following details: 
A) Rank
B) State
C) GSDP(18-19)- at current prices
D) GSDP(19-20)- at current prices
E) Share(18-19)
F) GDP($ billion)
Note: - From statisticstimes home page you have to reach to economy page through code.

```python
# Step 1: Set up the WebDriver
driver = webdriver.Chrome()
driver.get("https://statisticstimes.com/economy/india-statistics.php")

# Step 2: Navigate to the Economy Page
try:
    india_gdp_link = driver.find_element(By.XPATH, '/html/body/div[2]/div[2]/div[2]/ul/li[1]/a')
    india_gdp_link.click()
except NoSuchElementException:
    print("India GDP link not found")
    driver.quit()

# Step 4: Scrape the GDP data
time.sleep(3)
data = []
try:
    # Locate the table containing the data
    table = driver.find_element(By.XPATH, "//table[contains(@class,'display dataTable')]")
    rows = table.find_elements(By.TAG_NAME, "tr")

    # Loop through the table rows and extract data
    for row in rows[1:]:  # Skip the header row
        columns = row.find_elements(By.TAG_NAME, "td")
        if len(columns) > 0:
            rank = columns[0].text.strip()
            state = columns[1].text.strip()
            gsdp_21_22 = columns[4].text.strip()
            gsdp_22_23 = columns[3].text.strip()
            share_21_22 = columns[5].text.strip()
            gdp_billion_2021 = columns[6].text.strip()

            data.append([rank, state, gsdp_21_22, gsdp_22_23, share_21_22, gdp_billion_2021])
except NoSuchElementException:
    print("Table not found")
    driver.quit()

```

```python
# Step 5: Store the data in a pandas DataFrame
df = pd.DataFrame(data, columns={"Rank":rank, "State":state, "GSDP(21-22)":gsdp_21_22, "GSDP(22-23)":gsdp_22_23, "Share(21-22)":share_21_22, "GDP ($ billion)":gdp_billion_2021})
df


# Optional: Save to CSV
df.to_csv("state_gdp_india.csv", index=False)

# Step 6: Close the WebDriver
driver.quit()
```

# Question-4. Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used 
Note: - From the home page you have to click on the trending option from Explore menu through code. 

```python
# Step 1: Set up the WebDriver
driver = webdriver.Chrome() 
driver.get("https://github.com/explore")
```

```python
# Step 2: Click on the Explore menu and then on the Trending option
try:
    time.sleep(3)  # Wait for page to load
    # Click on the "Trending" option 
    trending_option = driver.find_element(By.XPATH, "/html/body/div[1]/div[4]/main/div[1]/nav/div/a[3]")
    trending_option.click()

except NoSuchElementException:
    print("Trending option not found")
    driver.quit()
```

```python
# Step 3: Scrape the trending repositories data
time.sleep(3)
repositories = []
try:
    # Locate the trending repositories section
    repo_list = driver.find_elements(By.XPATH, "//article[@class='Box-row']")

    for repo in repo_list:
        try:
            # Extracting repository title (Owner/Repository)
            repo_title = repo.find_element(By.XPATH, ".//h2[@class='h3 lh-condensed']/a").text.strip()
            
            # Extracting repository description
            try:
                repo_description = repo.find_element(By.XPATH, ".//p[@class='col-9 color-fg-muted my-1 pr-4']").text.strip()
            except NoSuchElementException:
                repo_description = "No description available"
            
            # Extracting contributors count
            try:
                contributors_count = repo.find_element(By.XPATH, ".//a[contains(@href, '/graphs/contributors')]").text.strip()
            except NoSuchElementException:
                contributors_count = "No contributors data"
            
            # Extracting programming language
            try:
                language = repo.find_element(By.XPATH, ".//span[@itemprop='programmingLanguage']").text.strip()
            except NoSuchElementException:
                language = "No language specified"
            
            # Append the data to the list
            repositories.append([repo_title, repo_description, contributors_count, language])
        
        except NoSuchElementException:
            continue

except NoSuchElementException:
    print("No repositories found on the page")
```

```python
# Step 4: Store the data in a pandas DataFrame
df = pd.DataFrame(repositories, columns = {"Repository Title":repo_title, "Repository Description":repo_description, "Contributors Count":contributors_count, "Language":language})
df

# Save to CSV
df.to_csv("trending_github_repositories.csv")


# Step 5: Close the WebDriver
#driver.quit()
```

## Question 5. Scrape the details of top 100 songs on billiboard.com. Url = https:/www.billboard.com/ 
You have to find the following details:

B) Artist name
C) Last week rank
D) Peak rank
E) Weeks on board

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import time

# Step 1: Set up the WebDriver
driver = webdriver.Chrome()
driver.get("https://www.billboard.com/charts/hot-100/")

# Step 2: Click on the Charts option and then Hot 100
try:
    time.sleep(3)  # Wait for page to load

    # Locate the "Charts" option
   # charts_option = driver.find_element(By.XPATH, "/html/body/div[3]/main/div[2]/div[1]/div[1]/div[1]/div[2]/div/div[1]/div/div[1]/div[1]/div/a")
    #charts_option.click()
    
    time.sleep(2)  # Small delay to ensure the menu is expanded
    
   

except NoSuchElementException:
    print("Charts or Hot 100 option not found")
    driver.quit()

# Step 3: Scrape the top 100 songs data
time.sleep(3)
songs = []
try:
    # Locate the section containing the songs
    song_list = driver.find_elements(By.XPATH, "//li[@class='o-chart-results-list__item']")

    for song in song_list:
        try:
            # Extract song name
            song_name = song.find_element(By.XPATH, "//h3[@class='c-title  a-no-trucate a-font-primary-bold-s u-letter-spacing-0021 u-font-size-23@tablet lrv-u-font-size-16 u-line-height-125 u-line-height-normal@mobile-max a-truncate-ellipsis u-max-width-245 u-max-width-230@tablet-only u-letter-spacing-0028@tablet']").text.strip()
            
            # Extract artist name
            try:
                artist_name = song.find_element(By.XPATH, ".//span[@class='c-label']").text.strip()
            except NoSuchElementException:
                artist_name = "Artist not available"
            
            # Extract last week rank
            try:
                last_week_rank = song.find_element(By.XPATH, ".//span[@class='c-label' and contains(text(), 'Last week')]").text.strip()
            except NoSuchElementException:
                last_week_rank = "N/A"
            
            # Extract peak rank
            try:
                peak_rank = song.find_element(By.XPATH, ".//span[@class='c-label' and contains(text(), 'Peak')]").text.strip()
            except NoSuchElementException:
                peak_rank = "N/A"
            
            # Extract weeks on board
            try:
                weeks_on_board = song.find_element(By.XPATH, ".//span[@class='c-label' and contains(text(), 'Weeks on chart')]").text.strip()
            except NoSuchElementException:
                weeks_on_board = "N/A"
            
            # Append the data to the list
            songs.append([song_name, artist_name, last_week_rank, peak_rank, weeks_on_board])

        except NoSuchElementException:
            continue

except NoSuchElementException:
    print("No songs found on the page")

# Step 4: Store the data in a pandas DataFrame
df = pd.DataFrame(songs, columns=["Song Name", "Artist Name", "Last Week Rank", "Peak Rank", "Weeks on Board"])

# Step 5: Display the DataFrame
print(df)

# Optional: Save the DataFrame to a CSV file
df.to_csv("billboard_top_100_songs.csv", index=False)

# Step 6: Close the WebDriver
#driver.quit()
```

## Question - 6: Scrape the details of Highest selling novels.
A) Book name
B) Author name
C) Volumes sold
D) Publisher
E) Genre
 Url - https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import time
# Step 1: Set up the WebDriver
driver = webdriver.Chrome()
driver.get("https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare")

# Step 2: Wait for the page to load completely
time.sleep(3)  # Wait for the page to fully load

# Step 3: Scrape the table containing the highest-selling novels data
novels = []
try:
    # Locate the table on the page
    table = driver.find_element(By.XPATH, "/html/body/div/div[2]/div[2]/div/div[2]/div/table")

    # Get all the rows in the table (excluding the header row)
    rows = table.find_elements(By.XPATH, ".//tbody/tr")

    for row in rows:
        # Extracting the columns from each row
        try:
            columns = row.find_elements(By.TAG_NAME, "td")
            if len(columns) > 0:
                # Extract the required data from each column
                book_name = columns[1].text.strip()
                author_name = columns[2].text.strip()
                volumes_sold = columns[3].text.strip()
                publisher = columns[4].text.strip()
                genre = columns[5].text.strip()

                # Append the data to the novels list
                novels.append([book_name, author_name, volumes_sold, publisher, genre])

        except NoSuchElementException:
            continue

except NoSuchElementException:
    print("Table or rows not found")

```

```python
# Step 4: Store the scraped data in a pandas DataFrame
df = pd.DataFrame(novels, columns={"Book Name":book_name, "Author Name":author_name, "Volumes Sold":volumes_sold, "Publisher":publisher, "Genre":genre})

# Step 5: Display the DataFrame
df

#Save the DataFrame to CSV
df.to_csv("highest_selling_novels.csv", index=False)

# Step 6: Close the WebDriver
driver.quit()
```

#### Question 2) Write a python program to scrape details of all the posts from https://www.patreon.com/coreyms .Scrape the heading, date, content and the likes for the video from the link for the youtube video from the post.


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL 
url = 'https://www.patreon.com/coreyms'

# Send GET request to the URL
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

posts = []

post_data = soup.find_all('div', class_='post-class')  

for post in post_data:
    heading = post.find('h2', class_='post-title-class').text.strip() 
    date = post.find('time', class_='post-date-class').text.strip()    
    content = post.find('div', class_='post-content-class').text.strip()  
    likes = post.find('span', class_='likes-class').text.strip()        
    video_link = post.find('a', class_='youtube-link-class')['href']   

    posts.append({
        'Heading': heading,
        'Date': date,
        'Content': content,
        'Likes': likes,
        'YouTube Link': video_link
    })

df = pd.DataFrame(posts)

# Display the DataFrame
print(df)

```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


url = 'https://www.patreon.com/coreyms'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

posts = soup.find_all('div', class_='post')

for post in posts:
  heading = post.find('h3').text
  date = post.find('time')['datetime']
  content = post.find('div', class_='post__content').text
  youtube_link = post.find('a', class_='post__youtube-link')['href']
  likes = post.find('span', class_='post__likes').text

  print(f'Heading: {heading}')
  print(f'Date: {date}')
  print(f'Content: {content}')
  print(f'YouTube Link: {youtube_link}')
  print(f'Likes: {likes}')
  print('\n')
```

#### Question-4) Write a python program to scrape first 10 product details which include product name , price , Image URL from
https://www.bewakoof.com/bestseller?sort=popular .

```python
page = requests.get('https://www.bewakoof.com/bestseller?sort=popular')
page
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL 
url = 'https://www.bewakoof.com/bestseller?sort=popular'

response = requests.get(url)

# Parse 
soup = BeautifulSoup(response.content, 'html.parser')

product_containers = soup.find_all('div', class_='productCardBox')[:10]

# Initialize a list 
products = []

for product in product_containers:
    # Extract the product name
    product_name = product.find('h3')
    product_name = product_name.text if product_name else 'N/A'
    
    # Extract the price
    price = product.find('span', class_='discountedPriceText')
    price = price.text if price else 'N/A'
    
    # Extract the image URL
    image = product.find('img')
    image_url = image['src'] if image else 'N/A'

    products.append({
        'Product Name': product_name,
        'Price': price,
        'Image URL': image_url
    })


df = pd.DataFrame(products)

# Print the DataFrame
print(df)
```

#### Question-5) Please visit https://www.cnbc.com/world/?region=world and scrap -
 a) headings
b) date
c) News link

```python
page = requests.get('https://www.cnbc.com/world/?region=world')
page
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


url = "https://www.cnbc.com/world/?region=world"
response = requests.get(url)

soup = BeautifulSoup(response.content, "html.parser")

articles = soup.find_all("div", class_="Card-titleContainer")


headlines = []
times = []
links = []

for article in articles:
    # Extract the headline
    headline_element = article.find("a")
    headline = headline_element.text.strip() if headline_element else None
    headlines.append(headline)
    
    # Extract the time
    time_element = article.find("time")
    time = time_element.text.strip() if time_element else 'No time available'
    times.append(time)
    
    # Extract the news link
    link = headline_element["href"] if headline_element else None
    links.append(link)


data = {
    "Headline": headlines,
    "Time": times,
    "News Link": links
}
df = pd.DataFrame(data)

# Print the df
print(df)
```
